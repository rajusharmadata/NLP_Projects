### . Import Libraries

```python
from sentence_transformers import SentenceTransformer, util
import torch
```

### Load Pretrained Model

```python
# This model converts sentences into embeddings
print("Loading model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!")
```

### Sentence Similarity

```python
sentence1 = "The cat is on the mat"
sentence2 = "Weather is so cold today"
```

```python
def sentence_similarity(sentence1: str, sentence2: str):
    """
    Compute similarity score between two sentences.
    """

    # Convert sentences into embeddings
    emb1 = model.encode(sentence1, convert_to_tensor=True)
    emb2 = model.encode(sentence2, convert_to_tensor=True)

    # Compute cosine similarity
    similarity = util.pytorch_cos_sim(emb1, emb2)

    return similarity.item()
```

### Paraphrase Mining

```python
def find_similar_sentences(sentences, top_k=5):
    """
    Finds similar sentence pairs from list.
    """

    # Generate embeddings
    embeddings = model.encode(sentences, convert_to_tensor=True)

    # Find similar sentence pairs
    paraphrases = util.paraphrase_mining_embeddings(embeddings)

    print("\nTop Similar Sentences:\n")

    for score, i, j in paraphrases[:top_k]:
        print(f"Sentence 1: {sentences[i]}")
        print(f"Sentence 2: {sentences[j]}")
        print(f"Similarity Score: {score:.4f}")
        print("-" * 50)
```

```python
import pandas as pd 
```

```python

# 1. Initialize a list to hold the formatted data
formatted_data = []

# 2. Iterate through the paraphrases
for paraphrase in paraphrases:
    score, i, j = paraphrase
    
    # Create a dictionary for the row, using formatting on the score
    row = {
        'Sentence 1': sentences[i],
        'Sentence 2': sentences[j],
        'Score': f"{score:.4f}" # Formatting score to 4 decimal places
    }
    formatted_data.append(row)

# 3. Create the DataFrame from the list of dictionaries
df = pd.DataFrame(formatted_data)

# 4. Display the DataFrame
df
```
