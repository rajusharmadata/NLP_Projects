```python
import warnings
warnings.filterwarnings('ignore')
import nltk
import string
nltk.download("punkt")
nltk.download("punkt_tab")
from nltk.tokenize import sent_tokenize,word_tokenize
```

### Lower Caseings

```python
text = """ Natural Languge Processing is fun to learn It helps computers to understand human languge !
Let's expolore tokenoization today
"""
```

```python
l_text = text.lower()

print("Lower case text :",l_text)
```

```python
sentences = sent_tokenize(l_text)
print("After Sentence Tokenize : \n",sentences,"Total Sentences are ","\n",len(sentences))
```

### Word Tokenization

```python
Words = word_tokenize(l_text)
print("After Word Tokenize : \n",Words,"Total Words are ","\n",len(Words))
```

```python
for i,s in enumerate(sentences,1):
      print(f"\n Sentences {i} : {s}")
      print("Token :",word_tokenize(s))
```
